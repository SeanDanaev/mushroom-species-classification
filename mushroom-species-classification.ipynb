###Downloading dataset and data understanding

```python
# Let's import all libraries we need
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, accuracy_score
from sklearn.metrics import ConfusionMatrixDisplay

url="https://raw.githubusercontent.com/alishermutalov/praktikum-datasets/refs/heads/praktikum/mushrooms.csv"
df = pd.read_csv(url)
df.head()
```

```python
#Find out about dataset
df.info()
```

```python
df.describe()
```

The dataset got 8124 rows and 23 columns. None NaN values found. DataTypes are all same "object". "veil-type" column is useless as we mentioned before and we are gonna drop it.

```python
#Get a copy of original dataset
df_mush = df.copy()
```

```python
# Drop not needed column
df_mush.drop("veil-type", axis=1, inplace=True)
```

```python
# Checking a column
df_mush["stalk-root"].value_counts()
```

```python
# "stalk-root" column's "?" value should be changed
df_mush["stalk-root"] = df_mush["stalk-root"].replace("?", np.nan)
df_mush["stalk-root"] = df_mush["stalk-root"].fillna(df_mush["stalk-root"].mode()[0])
```

```python
df_mush["stalk-root"].value_counts()
```

```python
# Let's encode columns

X = df_mush.drop('class', axis=1)
y = df_mush['class']

# Target -  y
le = LabelEncoder()
y = le.fit_transform(y)

# X
x_pipeline = ColumnTransformer(
    transformers=[
        ('ohe', OneHotEncoder(handle_unknown='ignore'), X.columns)
    ],
    remainder='drop'
)

X_encoded = x_pipeline.fit_transform(X)
```

###Data Analyzing

```python
# Vizualise all columns' data distribution
for col in df_mush.columns:
    plt.figure(figsize=(6,4))
    sns.countplot(data=df_mush, x=col)
    plt.title(f'{col} ustunining tarqalishi')
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
# Checking "class" column's values
df_mush["class"].value_counts()
```

```python
#Get encoded feature names
column_names = x_pipeline.get_feature_names_out()

#Convert sparse matrix to DataFrame
X_df = pd.DataFrame(X_encoded.toarray(), columns=column_names)

#Add the encoded y column
X_df['class'] = y

#Calculate correlation with only class column
corr_with_class = X_df.corr()['class'].sort_values(ascending=True)

#Convert to DataFrame for heatmap
corr_df = corr_with_class.to_frame()

plt.figure(figsize=(4, 20))   # Tall figure to see all features clearly
sns.heatmap(corr_df, annot=True, cmap='coolwarm')
plt.title("Correlation of Features with Class Label")
plt.show()
```

In the heatmap above, the negative values ​​are strongly associated with edibility (the 'e' value), while the positive values ​​show a strong association with toxicity (the 'p' value).

###Modelling and compare them

```python
# Train Test split
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
# KNN model's best parameters
param_grid = {"n_neighbors":range(1,31)}
knn = KNeighborsClassifier()
grid_search = GridSearchCV(knn,param_grid,cv=5,scoring="accuracy",n_jobs=-1)
grid_search.fit(X_train,y_train)
```

```python
print("Best k: ",grid_search.best_params_)
print("Best CV score: ", grid_search.best_score_)
```

```python
# KNN model
best_k = grid_search.best_params_["n_neighbors"]
knn_final = KNeighborsClassifier(n_neighbors=best_k)
knn_final.fit(X_train,y_train)
```

```python
# Predict
y_pred = knn_final.predict(X_test)
```

```python
# Model evaluation
cm_knn = confusion_matrix(y_test, y_pred)
precision_knn = precision_score(y_test, y_pred)
recall_knn = recall_score(y_test, y_pred)
f1_knn = f1_score(y_test, y_pred)
print("Confusion Matrix: \n", cm_knn)
print("Precision Score: ", precision_knn)
print("Recall Score: ", recall_knn)
print("F1 Score: ", f1_knn)
```

```python
# Decision Tree model's best parameters
param_grid = {
    "max_depth": range(1,21),
    "min_samples_leaf": range(10,101,10)
}
dt = DecisionTreeClassifier()
grid_dt = GridSearchCV(dt,param_grid,cv=5,scoring="accuracy",n_jobs=-1)
grid_dt.fit(X_train,y_train)
print("Best max_depth: ", grid_dt.best_params_)
print("Best CV sroce: ", grid_dt.best_score_)
```

```python
# Decision Tree model
best_depth = grid_dt.best_params_["max_depth"]
leaf_node = grid_dt.best_params_["min_samples_leaf"]
decision_tree = DecisionTreeClassifier(max_depth = best_depth,min_samples_leaf=leaf_node,random_state=42)
decision_tree.fit(X_train, y_train)
```

```python
# Predict from Tree model
y_pred_tree = decision_tree.predict(X_test)
```

```python
# Tree Model evaluation
cm_tree = confusion_matrix(y_test, y_pred_tree)
precision_tree = precision_score(y_test, y_pred_tree)
recall_tree = recall_score(y_test, y_pred_tree)
f1_tree = f1_score(y_test, y_pred_tree)
print("Confusion Matrix: \n", cm_tree)
print("Precision Score: ", precision_tree)
print("Recall Score: ", recall_tree)
print("F1 Score: ", f1_tree)
```

```python
# Let's visualize our Tree model
cols = X_df.drop("class",axis=1).columns
classes = df_mush["class"].unique()
plt.figure(figsize=(20,20))
tree.plot_tree(decision_tree,feature_names=cols,class_names=classes,filled=True)
plt.show()
```

```python
# RandomForest model's best parameters
param_grid = {
    "n_estimators":[100,200],
    "max_depth": [None,10,20],
    "min_samples_leaf": [5,10,15],
    "max_features": ["sqrt", "log2"]
}

grid_rf = GridSearchCV(RandomForestClassifier(random_state=42),param_grid,cv=5,scoring="accuracy",n_jobs=-1)
grid_rf.fit(X_train,y_train)
```

```python
print("Best parameters : ", grid_rf.best_params_)
print("Best score: ", grid_rf.best_score_)
```

```python
# Random Forest model
best_params = grid_rf.best_params_
rf_final = RandomForestClassifier(**best_params,random_state=42)
rf_final.fit(X_train,y_train)
```

```python
# get prediction from Random Forest model
y_pred_rf = rf_final.predict(X_test)
```

```python
# Model evaluation for Random Forest
cm_rf = confusion_matrix(y_test, y_pred_rf)
precision_rf = precision_score(y_test, y_pred_rf)
recall_rf = recall_score(y_test, y_pred_rf)
f1_rf = f1_score(y_test, y_pred_rf)
print("Confusion Matrix: \n", cm_rf)
print("Precision Score: ", precision_rf)
print("Recall Score: ", recall_rf)
print("F1 Score: ", f1_rf)
```

```python
# Logistic Regression Classification
log_reg = Pipeline([
    ("scaler", StandardScaler(with_mean=False)),
    ("lr", LogisticRegression(max_iter=500))
])
log_reg.fit(X_train, y_train)
```

```python
# Logistic Regression model predictions
y_pred_log_reg = log_reg.predict(X_test)
```

```python
# Model evaluation
accuracy_log = accuracy_score(y_test,y_pred_log_reg)
cm_log = confusion_matrix(y_test, y_pred_log_reg)
precision_log = precision_score(y_test, y_pred_log_reg)
recall_log = recall_score(y_test, y_pred_log_reg)
f1_log = f1_score(y_test, y_pred_log_reg)
print("Accuracy : ", accuracy_log)
print("Confusion Matrix: \n", cm_log)
print("Precision Score: ", precision_log)
print("Recall Score: ", recall_log)
print("F1 Score: ", f1_log)
```

###Visuals of all models' performances

```python
# Confusion Matrix for all models
models = {"KNN":knn_final,"Decision Tree":decision_tree,"Random Forest":rf_final,"Logistic Regression":log_reg}

plt.figure(figsize=(12,12))
for name, model in models.items():
  y_pred = model.predict(X_test)
  disp = ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
  plt.title(f"Confusion Matrix of {name}")
  plt.show()
```

```python
# Let's plot all models' outcome as a bargraph
models = {"KNN":knn_final,"Decision Tree":decision_tree,"Random Forest":rf_final,"Logistic Regression":log_reg}

precision_list = []
recall_list = []
f1_list = []

for name , model in models.items():
  y_pred = model.predict(X_test)
  precision_list.append(precision_score(y_test,y_pred))
  recall_list.append(recall_score(y_test,y_pred))
  f1_list.append(f1_score(y_test,y_pred))
```

```python
x = np.arange(len(models))
width = 0.25

fig, ax = plt.subplots(figsize=(10,6))
ax.bar(x - width,precision_list,width,label="Precision")
ax.bar(x,recall_list,width,label="Recall")
ax.bar(x + width,f1_list,width,label="F1")

ax.set_xticks(x)
ax.set_xticklabels(models.keys())
ax.set_ylim(0, 1.05)
ax.set_ylabel("Score")
ax.set_title("Hamma Modellarning score lari solishtirmasi")
ax.legend()

plt.show()
```

###Final Insight
